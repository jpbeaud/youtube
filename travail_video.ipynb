```python
from moviepy.editor import VideoFileClip, ImageClip

# Charger la vidéo
video = VideoFileClip("Ethiopia Runners.mp4")

# Charger l'image à superposer
image = ImageClip("/content/IMG20240224075204.jpg").set_duration(video.duration)

#Resize the image to the size of the video
# This will ensure the image and video have the same dimensions for the mask to work.
image = image.resize(video.size)

# Convertir l'image en masque (noir et blanc)
mask = image.to_mask()

# Superposer l'image sur la vidéo en utilisant le masque
final_clip = video.set_mask(mask)

image.write_videofile("image_superposee.mp4",fps=24)
f_clip = VideoFileClip("image_superposee.mp4")
f_clip.ipython_display(width = 280)
# Enregistrer le résultat
#final_clip.write_videofile("video_avec_image.mp4")

#final_clip.ipython_display(width = 280)
```
